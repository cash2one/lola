```python
import requests
import bs4
```

```python
quotes_root_page = "http://leagueoflegends.wikia.com/wiki/Category:Champion_quotes"
base_url = "http://leagueoflegends.wikia.com/wiki/"
response = requests.get(quotes_root_page)
soup = bs4.BeautifulSoup(response.text)
#print soup
```

```python
main_content = soup.find_all('div', class_='mw-content-ltr')
all_items = main_content[0].find_all('li')
all_links = [each_item.find('a')['title'] for each_item in all_items]
print all_links
```

```python
f = open('champions_quotes.txt', 'w')
for each_link in all_links:
    f.write(each_link + '\n')
    print each_link
    full_url = base_url + each_link
    response = requests.get(full_url)
    soup = bs4.BeautifulSoup(response.text)
    all_audios = soup.find_all('li')
    for a_audio in all_audios:
        a_quote = a_audio.find('i')
        if a_quote != None:
            #print a_quote.text
            f.write(a_quote.text.encode('utf-8') + '\n')
```
