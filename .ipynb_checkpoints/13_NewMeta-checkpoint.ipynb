```python
def GetWinningSide(a_match):
    winning_side = -1
    if a_match['teams'][0]['winner'] == True and a_match['teams'][1]['winner'] == False:
        winning_side = 0
    if a_match['teams'][0]['winner'] == False and a_match['teams'][1]['winner'] == True:
        winning_side = 1 
    return winning_side
```

```python
def GetTierPoint(tier, division):
    point = 0
    if tier == 'SILVER':
        point = 10
    if tier == 'GOLD':
        point = 20
    if tier == 'PLATINUM':
        point = 30
    if tier == 'DIAMOND':
        point = 40
    if tier == 'MASTER':
        point = 50
    if tier == 'CHALLENGER':
        point = 60
    
    if division == 'IV':
        point += 2
    if division == 'III':
        point += 4
    if division == 'II':
        point += 6
    if division == 'I':
        point += 8
        
    return point
```

```python
def AddLeaguesToDB(leagues):
    current_leaguedb = None
    for a_league in leagues:
        a_league_data = json.loads(a_league.to_json())
        league_name = a_league_data['name']
        if league_names.find_one({'leagueName':league_name}) == None:
            print 'Add league:' + league_name
            league_names.insert_one({'leagueName':league_name})
            league_queue = a_league_data['queue']
            if league_queue == 'RANKED_SOLO_5x5':
                current_leaguedb = solo_leagues
            elif league_queue == 'RANKED_FLEX_SR':
                current_leaguedb = flex_leagues
            else:
                print "MISSING LEAGUE:" + league_queue
                break
            league_tier = a_league_data['tier']
            league_entries = a_league_data['entries']
            for a_entry in league_entries:
                league_division = a_entry['division']
                a_point = GetTierPoint(league_tier, league_division)
                summonerId = a_entry['playerOrTeamId']
                if current_leaguedb.find_one({'summonerId':summonerId}) == None: 
                    current_leaguedb.insert_one({'summonerId':summonerId, 'point':a_point})
```
